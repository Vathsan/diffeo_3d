```python
def display_im_grid(xgrid, ygrid, im, ngrid):
    """Display both grids and image"""
    clear_output(wait=True)
    plt.imshow(im, cmap="gray")
    plt.plot(ygrid[::ngrid, ::ngrid], xgrid[::ngrid, ::ngrid], "b", lw=1.0)
    plt.plot(ygrid[::ngrid, ::ngrid].T, xgrid[::ngrid, ::ngrid].T, "b", lw=1.0)

    plt.axis("off")
    plt.axis("equal")
    plt.pause(0.1)
```

```python
im = np.zeros((81, 121))

# Create a grid image
im[::5, :], im[:, ::5] = 1, 1
im[::5, ::5] = 1


szx, szy = im.shape

nframes = 10
j_lb = 0.5
j_ub = 2.0

# Generate radial and rotational fields
f11 = np.linspace(1.0, 0.3, nframes)
f12 = np.linspace(1.0, 3.0, nframes)
f21 = np.linspace(0.0, 1.0, nframes)
f22 = np.linspace(0.0, -1.0, nframes)

cz = int(0.1 * min(szx, szy))

# radial field
f1 = np.ones((szx,szy))

# rotational field
f2 = np.zeros((szx,szy))

for i in range(nframes):
    
    f1[
        szx // 3 - 1 - cz : szx // 3 + cz, szy // 3 - 1 - cz : szy // 3 + cz
    ] = f11[i]
    f1[
        2 * szx // 3 - 1 - cz : 2 * szx // 3 + cz,
        2 * szy // 3 - 1 - cz : 2 * szy // 3 + cz,
    ] = f12[i]
    f2[
        szx // 3 - 1 - cz : szx // 3 + cz, szy // 3 - 1 - cz : szy // 3 + cz
    ] = f21[i]
    f2[
        2 * szx // 3 - 1 - cz : 2 * szx // 3 + cz,
        2 * szy // 3 - 1 - cz : 2 * szy // 3 + cz,
    ] = f22[i]    
    
    posx, posy = gridgen(f1, f2, j_lb, j_ub, inv=False)
    posx_inv, posy_inv = gridgen(f1, f2, j_lb, j_ub, inv=True)
    
    imw = mygriddata(posx_inv, posy_inv, im)
    
    display_im_grid(posx, posy, imw, 5)
```
